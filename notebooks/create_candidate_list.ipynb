```python
def get_authors_and_abstracts(refs):
    ids = [w["id"].split("/")[-1] for w in refs]
    authorships = [w["authorships"] for w in refs]
    authors = [j[0]["author"] for j in authorships]
    abstracts = [w["abstract"] for w in refs]

    return dict(zip(ids, zip(authors, abstracts)))


get_authors_and_abstracts(refs) 
```
